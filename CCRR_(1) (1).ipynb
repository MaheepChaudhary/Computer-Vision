```python
import random
l,p,unchanged_article_list=[],[],[]
start=0
genre=[]
#l is list of user
#p is list of articles,unchanged_article_list is same as p but reserved for dataframe 
for i in range(1,101):
    l.append("user"+""+str(i)) 
for j in range(1,201):
    p.append("article "+str(j))
    unchanged_article_list.append("article "+str(j))

# here we are dividing artices into genres
'''
for j in range(1,11):
    for k in range(start,20*j):
        p[k] += (" genre "+str(j))
    start=20*int(j)
'''
    
#making a genre folder with 20 article in every genre
k=0
for i in range(1,11):
    genre.append(p[k:k+20])
    k+=20

#Now dividing list of users into 3 categories randomly
random.shuffle(l)
beginner = l[0:30]
medium = l[30:80]
expert = l[80:]

#gen_beg is the genres asiigned to beginner user 
gen_beg = []
for i in range(30):
    x=[]
    n = random.randint(5,7)
    x=(random.sample(range(1,10),n))
    gen_beg.append(x)    


#gen_med is the genres asiigned to medium user 
gen_med = []
for i in range(50):
    x=[]
    n = random.randint(3,4)
    x=(random.sample(range(1,10),n))
    gen_med.append(x)    

#gen_exp is the genres assigned to expert user 
gen_exp = []
for i in range(20):
    x=[]
    n = random.randint(1,2)
    x=(random.sample(range(1,10),n))
    gen_exp.append((x))    
print(p)
```

```python
#for all beginner code and begdict is the dictionary which holds the user and coresponding addresed articles with genres
begdict={}
for i in range(len(beginner)):
    n=[]
    for k1 in gen_beg[i]:
        random.shuffle(genre[(int(k1)-1)])
        num=random.randint(2,4)
        n.extend(genre[(int(k1)-1)][:num])
    begdict[beginner[i]]=n
    
for keys,values in begdict.items():
    for q in range(len(values)):
        begdict[keys][q] = tuple(list([begdict[keys][q],random.randint(3,5)]))
       
print(len(begdict))
print(len(beginner))
```

```python
#for all medium code and meddict is the dictionary which holds the user and coresponding addresed articles with genres
meddict={}
for i in range(len(medium)):
    n=[]
    for k1 in gen_med[i]:
        random.shuffle(genre[(int(k1)-1)])
        num=random.randint(6,10)
        n.extend(genre[(int(k1)-1)][:num])
    meddict[medium[i]]=n
    
for keys,values in meddict.items():
    for q in range(len(values)):
        meddict[keys][q] = tuple(list([meddict[keys][q],random.randint(3,5)]))

print(len(meddict))
print(len(medium))
```

```python
#for all expert code and expdict is the dictionary which holds the user and coresponding addresed articles with genres
expdict={}
for i in range(len(expert)):
    n=[]
    for k1 in gen_exp[i]:
        random.shuffle(genre[(int(k1)-1)])
        num=random.randint(12,16)
        n.extend(genre[(int(k1)-1)][:num])
    expdict[expert[i]]=n

for keys,values in expdict.items():
    for q in range(len(values)):
        expdict[keys][q] = tuple(list([expdict[keys][q],random.randint(3,5)]))

print(len(expdict))
print(len(expert))
```

```python
# beg_article_ratings is a list of articles with their respective ratings

beg_article_ratings = []

for i in beginner:
    temp = [0]*200
    for j in begdict[i]:
        temp[int(j[0][7:])-1] = int(j[1])
    beg_article_ratings.append(temp)
    
print(len(beg_article_ratings))
print(len(beginner))
        
        
```
