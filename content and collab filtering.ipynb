```python
import pandas as pd
import numpy as np
import os

df1 = pd.read_csv("car.csv", error_bad_lines=False)
df2 = pd.read_csv("JAE.csv", error_bad_lines=False,encoding = 'unicode_escape')
df3 = pd.read_csv("JAR.csv", error_bad_lines=False)
df3=df3.rename(columns={'article_name ':'article_name'})
df = pd.concat([df1, df2, df3], axis=1)
df.to_csv('df.csv')
a = list(set(df.columns))
#print(a)
resu = pd.DataFrame(columns=a)
Res=pd.concat([df1,resu,df2,df3],axis=0)
Res.to_csv('Res.csv')
pd.set_option('display.max_columns',25)
#Res.set_index('article_name',inplace=True)
Res = Res.replace(np.nan,"",regex=True)
Res['author'] =Res['author1'] + "," + Res['author2'] +"," +  Res['author4'] +"," + Res['author3'] +"," + Res['authors']+"," + Res['article_authors']
Res['institute']=Res['institute1'] +',' + Res['institute2']+ ',' + Res['institute3'] + "," + Res['institute4']
Res['volume']=Res['volume_issue'] +',' + Res['volumes']+ ',' + Res['subvolumes'] + "," + Res['subvolumes-href']
Res["Date"] = Res['publication_date']+','+Res['publication_year']
Res.drop(columns=['author3', 'volumes', 'article_title', 'article_authors', 'institute4', 'publication_year', 'version', 'institute3', 'institute2', 'author4', 'publication_date', 'subvolumes', 'institute1', 'subvolumes-href', 'author2', 'volume_issue', 'author1', 'authors', 'volumes-href'],inplace=True)
#Res.head(3)
Tf=Res
Tf['colm']=Tf['article_name']+','+ Tf['journal_name']+','+Tf['author']+","+Tf['institute']+ ","+Tf['volume']+','+Tf['Date']
Tf.drop(columns=['article_name','journal_name','author','institute','volume','Date'],inplace=True)
Tf.head(3)
```

```python
import re, math
from collections import Counter
a=np.zeros([2860,21])
#print(a)

WORD = re.compile(r'\w+')

def get_cosine(vec1, vec2):
     intersection = set(vec1.keys()) & set(vec2.keys())
     numerator = sum([vec1[x] * vec2[x] for x in intersection])

     sum1 = sum([vec1[x]**2 for x in vec1.keys()])
     sum2 = sum([vec2[x]**2 for x in vec2.keys()])
     denominator = math.sqrt(sum1) * math.sqrt(sum2)

     if not denominator:
        return 0.0
     else:
        return float(numerator) / denominator

def text_to_vector(Text):
     words = WORD.findall(Text)
     return Counter(words)
for i in range(0,2860):
    for j in range(0,21):
        vector1 = text_to_vector(Tf.iloc[i,-1])
        vector2 = text_to_vector(Tf.iloc[j,-1])
        a[i][j] = get_cosine(vector1, vector2)

print(a)
```
