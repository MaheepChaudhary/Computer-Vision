```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

road = cv.imread("road.png")

def ROI(canny_image,vertices):
    mask = np.zeros_like(canny_image)
    color = 255
    mask = cv.fillPoly(mask,vertices,color)
    final = cv.bitwise_and(canny_image,mask)
    return final

def draw_lines(lines,image):
    image = np.copy(image)
    blank_image = np.zeros((image.shape[0],image.shape[1],3),dtype = np.uint8)
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv.line(blank_image,(x1,y1),(x2,y2),(0,255,0),thickness=3)
            print(image.shape)
            print(blank_image.shape)
        final = cv.addWeighted(image,0.8,blank_image,1,0.0)
    return final
        
def process(frame):
    gray = frame.copy()
    im = cv.cvtColor(gray,cv.COLOR_BGR2GRAY)
#    im = cv.blur(im,(5,5))
    canny_img = cv.Canny(im,200,250)
    height = im.shape[0]
    width = im.shape[1]
    vertices = [(0,height),(600,500),(width,height)]
    cropped_image = ROI(canny_img,np.array([vertices],np.int32)) 
    lines = cv.HoughLinesP(cropped_image, rho=6, theta=np.pi/180, lines=np.array([]),threshold=160, minLineLength=40, maxLineGap=27)
    image_with_lines = draw_lines(lines,frame)
    return image_with_lines
    
vid = cv.VideoCapture("road.mp4")   
while(vid.isOpened()):
    ret,frame = vid.read()
#    frame = cv.cvtColor(frame,cv.COLOR_BGR2)
    imf = process(frame)
    cv.imshow("image",imf)
    if cv.waitKey(1) & 0xFF == ord("q"):
        break
vid.release()        
cv.destroyAllWindows()
```
